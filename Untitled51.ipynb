```python
import numpy as np
def f(x):
    return np.sin(x) - np.log(x) + np.exp(x)
a = 0.2
b = 1.4
h = 0.15
x = np.arange(a, b + h, h)
y = f(x)

print("x values:", x)
print("f(x) values:", y)
t=(h/2)*(y[0]+2*(y[1]+y[2]+y[3]+y[4]+y[5]+y[6]+y[7])+y[8])
s1=(h/3)*(y[0]+4*(y[1]+y[3]+y[5]+y[7])+2*(y[2]+y[4]+y[6])+y[8])
s2=(3*h/8)*(y[0]+3*(y[1]+y[2]+y[4]+y[5]+y[7])+2*(y[3]+y[6])+y[8])
print("Trapezoidal value:", t)
print("Simpson's one-third value:", s1)
print("Simpson's three-eight value:", s2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'bo-', label='f(x)')
plt.fill_between(x, y, alpha=0.3, color='skyblue')
plt.title('Area of the function')
plt.xlabel('Value of x')
plt.ylabel('Value of f(x)')
plt.legend()
plt.show()
```

```python
import numpy as np
def f(x):
    return 1/(1+(x**2))
a = 0
b = 1
h = 0.001
x = np.arange(a, b + h, h)
y = f(x)
n=len(x)
print("length:", n)
print("x values:", x)
print("f(x) values:", y)
t1=0
for i in range (1,n-1, 1):
    t1 += y[i]
trap=(h/2)*(y[0]+2*t1+y[n-1])
print("Trapezoidal value:", trap)
```

```python
s1=0
s2=0
for i in range (1,n-1, 2):
    s1 += y[i]
for i in range (2,n-2, 2):
    s2 += y[i]
sim1=(h/3)*(y[0]+4*s1+2*s2+y[n-1])
print("Simpson's one-third value:", sim1)
```

```python
s3 = 0
s4 = 0
for i in range(1, n - 1):
    if i % 3 == 0:
        s4 += y[i]
    else:
        s3 += y[i]
sim3 = (3 * h / 8) * (y[0] + 3 * s3 + 2 * s4 + y[n - 1])
print("Simpson's three-eighth value:", sim3)
```
